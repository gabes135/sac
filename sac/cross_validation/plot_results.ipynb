# Load in the process validation $\chi^2$ files

```python
folder = 'bosonic_delta_peak'
```

```python
val_free = pd.read_csv(f'out_files/{folder}/chi2/free_v.csv')
val_peak = pd.read_csv(f'out_files/{folder}/chi2/peak_v.csv')
samp_free = pd.read_csv(f'out_files/{folder}/chi2/free_s.csv')
samp_peak = pd.read_csv(f'out_files/{folder}/chi2/peak_s.csv')

print("Sampling chi^2 minimum for unconstrained parameterization:
samp_free.mean().min()
```

```python
theta = val_free.columns.astype(float)
plt.plot(theta, val_free.mean(), label = 'Unconstrained')
plt.plot(theta, val_peak.mean(), label = r'Delta Peak, $a_0 = 0.70$')

plt.xscale("log")
plt.ylim(0.75, 2)
plt.legend()
plt.xlabel("$\Theta$")
plt.ylabel(r"$\langle \chi^2_{\mathrm{val.}} \rangle/N_{\tau}$")
```

```python
spec_peak = pd.read_csv("../peak/out_files/bosonic/Np_01/A0_0.700/sw000.csv")
spec_free = pd.read_csv("../free/out_files/bosonic_delta_peak/sw000.csv")
spec_true = pd.read_csv("../in_files/peak/aw_b.dat")
```

```python
plt.plot(spec_true.omega, spec_true.S, c= 'k', lw = 2, label = 'True Spectrum')
plt.plot(spec_free.omega, spec_free.S, label = 'Unconstrained')
plt.plot(spec_peak.omega, spec_peak.S, label =  r'Delta Peak, $a_0 = 0.70$')

plt.legend()

plt.xlabel("$\omega$")
plt.ylabel("$S(\omega)$")

plt.ylim(0, 7)
plt.xlim(0, 3)
```
