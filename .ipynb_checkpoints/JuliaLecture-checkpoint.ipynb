### **Functions**

```julia
# one line function definition (nice for simple math)
gauss(x,μ,σ) = 1/(σ*√(2π)) * exp(-1/2 * ((x-μ)/σ)^2)

# recursion, specifying type (we'll come back to the relationship between functions and input types)
factorial(n::Integer) = n <= 1 ? 1 : n*factorial(n-1)

# multiline function definition
function myexp(x;ntosum = 20)
    total = zero(x)
    for n in 0:ntosum
        total += x^n / factorial(n)
    end
    return total
end
```

### **Don't fear `for` loops**

```julia
function gausssum(μ,σ)
    total = 0.0
    for i in -1000:0.001:1000
        total += gauss(i,μ,σ)
    end
    return total*0.001
end

@benchmark gausssum(0,1)
```

Here is a version with all array operations, it is basically the same in terms of efficiency. 

```julia
@. gaussbroadcast(x,μ,σ) = 1/(σ*√(2*π)) * exp(-1/2 * ((x-μ)/σ)^2)

@benchmark sum(gaussbroadcast(Array(-1000:0.001:1000),0,1))*0.001
```

## **3. Key Paradigm: Multiple Dispatch**

#### *All named functions are generic* 

### **Multiple Dispatch is Unreasonably Effective**

**A function represents an abstract mathematical operation - it can operate differently on different input types. This is handled during compiling - doesn't effect runtime after just in time compilation\*.**

**More flexible than single dispact, eg `Type.method(self,args..)`. With single dispatch, methods are inside of types. What if you want to define a new method for an existing type?**
- Modify the source code?...Burdens the original author. Who decides which methods should be included in the class?
- Write a new subclass / inherit?...Code written with the old type can't use the new method. We didn't want a whole new thing!
- Fork the code, modify the class?...Sure, but now you can't share easily. 
- Write functions rather than methods?...Not generic, requires different functions for different types.  

Note: Function overloading is not the same as multiple dispatch...it replaces the actual type with the static type. See for example https://youtu.be/kc9HwsxE1OY?t=399

\*There are exceptions where this doesn't work, non-constant global variables and ambigiously defined types for example. Don't use these in performance sensitive code.  


### **Example 1: Vectors, Matrices, and the Kronecker $\delta$**

**Here are a few types of vectors which can use the same operations.**

```julia
using SparseArrays, StaticArrays, LinearAlgebra,BenchmarkTools

function matrixelement(v1::AbstractVector,M::AbstractMatrix,v2::AbstractVector)
    return dot(v1,M*v2)
end

L = 100
v = sin.(1:L)
M = rand(L,L)
Mstatic = SMatrix{L,L}(M)
vstatic = SVector{L,Float64}((sin.(1:L)))
nnonzero = 10
vsparse = sparsevec(Int64.(ceil.(rand(nnonzero)*L)), rand(nnonzero),L)

println(dot(v,vstatic))
println(dot(v,vsparse))
matrixelement(v,M,v)
matrixelement(vstatic,M,vsparse)
@btime matrixelement(v,M,v)
@btime matrixelement(vstatic,Mstatic,vstatic)
@btime matrixelement(vsparse,M,vsparse)
```

```julia
struct kroneckerδ <: AbstractVector{Bool}
    i::Int64 # the non-zero index
    L::Int64 # the length
end 

import Base: length,getindex,size

length(v::kroneckerδ) = v.L
size(v::kroneckerδ) = (v.L,)
getindex(v::kroneckerδ,i::Int64) = i == v.i

δ3 = kroneckerδ(3,L)
δ6 = kroneckerδ(6,L)

@btime matrixelement(δ6,M,δ3) # only slightly faster than with normal vectors
```

**Inner products with the Kronecker $\delta$ is only slightly faster than with normal vectors, because it's falling back to methods for `AbstractVector`**  
**Solution: Write a better dot product and matrix multiplication specialized for Kronecker $\delta$'s**

```julia
import LinearAlgebra.dot
dot(v1::kroneckerδ,v2::AbstractVector) = v2[v1.i]
*(A::AbstractMatrix,v::kroneckerδ) = A[:,v.i]

@btime matrixelement(δ6,M,δ3) # about a factor of 10 faster
```

**But we can do even better by writing a new `matrixelement` method**

```julia
matrixelement(v1::kroneckerδ,M::AbstractMatrix,v2::kroneckerδ) = M[v1.i,v2.i]

@btime matrixelement(δ6,M,δ3) # 70 x faster than the original 
```

**Multiple dispatch can sometimes identify speed up a program without any input from the programmer. Here, it identifies symmetries of a matrix to speed diagonalization.**

```julia
A = rand(100,100)
Asym = (A + A')/2
Asymexplicit = Symmetric((A+A')/2)
Aalmostsym = (A + 0.99999999*A')/2

println("Time to diagonalize")
@btime eigen(A);
println("Time to diagonalize Asym")
@btime eigen(Asym);
println("Time to diagonalize Asym explicitly marked as symmetric")
@btime eigen(Asymexplicit)
println("Time to diagonalize Aalmostsym")
@btime eigen(Aalmostsym);
```

### **Example 2: `Measurements.jl` + `DifferentialEquations.jl` = Error Propagation through DiffEqs with no extra work**
#### Multiple dispatch makes julia very sharable!


```julia
using DifferentialEquations,Measurements,Plots
```

```julia
g = 9.8 ± 0.02 # gravity
l = 1.0 ± 0.02 # length of pendulum
u0 = [0 ± 0, pi / 60 ± 0.0] # initial angle, angular velocity
tspan = (0 ± 0,6.3 ± 0) 

pendulum(u,p,t) = [u[2],-(g/l)*u[1]]

# use ODE solvers
prob = ODEProblem(pendulum,u0,tspan);
sol = solve(prob,Tsit5(),reltol = 1e-6);

# Analytic solution 
# u = u0[2] .* cos.(sqrt(g/l) .* sol.t);
```

```julia
plot(sol.t,[i[1] for i in sol.u],ylabel = "θ",xlabel = "t",label = nothing,title = "Uncertain Frequency",yticks = [-0.02,-0.01,0.0,0.01,0.02])
```

```julia
g2 = 9.8 ± 0.0 # gravity
l2 = 1.0 ± 0.0 # length of pendulum
u02 = [0 ± 0, pi / 60 ± 0.01] # initial angle, angular velocity

pendulum2(u,p,t) = [u[2],-(g2/l2)*u[1]]

# use ODE solvers
prob2 = ODEProblem(pendulum2,u02,tspan);
sol2 = solve(prob2,Tsit5(),reltol = 1e-6);

plot(sol2.t,[i[1] for i in sol2.u],ylabel = "θ",xlabel = "t",label = nothing,title = "Uncertain Amplitude")
```

### **Example 3: Simple Automatic Differentiation with Multiple Dispatch**

We will define dual numbers, $x + \delta x$, in which any value of $\mathcal{O}(\delta x^2) = 0$. By imlementing simple operations (+,-,\*,/,^,etc) for dual numbers, we can efficiently compute derivatives of nearly any julia program.

Here are some good overviews of Automatic Differentiation, including reverse mode AD, which is usually more efficient for gradients https://www.youtube.com/watch?v=wG_nF1awSSY, https://www.youtube.com/watch?v=FtnkqIsfNQc. 

```julia
struct dual{T<:Union{Float64,ComplexF64}} <: Number
    x :: T
    δx :: T
end 

import Base: +,-,*,/,^,convert, promote_rule,exp,abs2,sqrt,sin,cos
+(x1::dual,x2::dual) = dual(x1.x+x2.x,x1.δx + x2.δx)
*(x1::dual,x2::dual) = dual(x1.x*x2.x,x1.x*x2.δx + x2.x*x1.δx)
-(x1::dual,x2::dual) = dual(x1.x-x2.x,x1.δx - x2.δx)
/(x1::dual,x2::dual) = dual(x1.x/x2.x,(x2.x*x1.δx - x1.x*x2.δx) / x2.x^2)
^(x1::dual,x2::dual) = dual(x1.x^x2.x,x2.x*x1.x^(x2.x-1)*x1.δx + x1.x^x2.x*log(x1.x)*x2.δx)

convert(::Type{dual{Float64}},x::Real) = dual(Float64(x),0.0)
convert(::Type{dual{ComplexF64}},x::Union{Complex,Real}) = dual(ComplexF64(x),0.0+0.0im)
convert(::Type{dual{ComplexF64}},x::dual{Float64}) = dual(ComplexF64(x.x),ComplexF64(x.δx))
promote_rule(::Type{dual{Float64}},::Type{<:Real}) = dual{Float64}
promote_rule(::Type{dual{Float64}},::Type{<:Complex}) = dual{ComplexF64}
promote_rule(::Type{dual{ComplexF64}},::Type{<:Number}) = dual{ComplexF64}
#promote_rule(::Type{<:Number},::Type{dual}) = dual
#promote_rule(::Type{<:Number},::Type{dual}) = dual
exp(x1::dual) = dual(exp(x1.x),exp(x1.x)*x1.δx)
abs2(x1::dual) = dual(abs(x1.x^2),abs(2*x1.δx*x1.x))
sqrt(x1::dual) = x1^(1/2)
-(x1::dual) = -1*x1
```

```julia
x = dual(5.0,1.0)
y = dual(3.0,1.0)

f(x) = x^4 + 5*x^3 - 3*x + 7
fprime(x) = 4*x^3 + 15*x^2 - 3

println(f(10.0)," ",fprime(10.0))
println(f(dual(10.0,1.0)))
```

```julia
using LinearAlgebra

import LinearAlgebra.eigsortby
dual{Float64}(x::Bool) = dual(Float64(x),0.0)
eigsortby(x::dual{Float64}) = eigsortby(x.x)
eigsortby(x::dual{ComplexF64}) = eigsortby(x.x)

println(typeof(x*[1.0 0.0 ; 0.0 1.0]))
println(typeof([x+y y;y x-y]))
```

### **A. Tips for using Julia**
- Don't use global variables in performace sensitive code (use constants instead). 
- Type declarations are not necessary for efficient code. Declare types for fields in `struct`s, when a function will operate differently on different types, or when it significantly benefits clarity.
- Write "type stable" functions, meaning the output type depends only on the input type (but not the input value).
- To write general code, just leave out the types :) (this is a bit of an over-simplification but true in many cases)
- Expect some compile time. The first time a function is called with take longer (so called "time to first plot" problem). 
- You can use `PyCall` to run python code in julia, if you must. Similar methods exist for other languages.
- Type `?` in the repl for documentation, eg `? eigen`
- General style rules are that types and modules should be capitalized, variables and functions should be lowercase. Functions which mutate should end in `!`.

### **D. Myths About Julia**
- *You need to declare types for julia to be fast*: This is not true, julia will compile code optimized for the types on which a function is called. It is however important to respect type stability and avoid untyped global variables. 
- *Julia documentation is far worse than python*: This is not true in my experience. While there are fewer people discussing julia on stackoverflow, the base documentation is very good. For 3rd party repositories, the quality of documentation obviously varies, but in general julia developers like sharable, high quality, scientific programming. Also, julia often lends itself to more general and intuitve code than other languages.
- *Julia is unlikely to have the specialized tools I need*: The number of julia users and repositories has been growing exponentially for around a decade, and they are concentrated in numerical scientific computing. Of course, julia will not have every tool in existence, but this can sometimes be fixed by calling code from other languages. 
- *Julia is just fast python OR I can get the same benefits from numpy / cython*: numpy / cython can be high performance if everything is vectorized, but sometimes it is not obvious how to do this properly (eg `np.vectorize` is apparently just a for loop, while broadcasting in julia works nicely). However, performance aside, julia paradigms like multiple dispatch often lead to more intuitive, general, and sharable code than python paradigms. 
- *Julia is super niche*: This is really not true anymore. It is probably the fourth most popular language for high performance numerical computing (after C, C++, and Fortran) and is certainly in the top 5-10 used in science. 

### **E. Why NOT Julia**
- If you need the script to run exceptionally quickly (milliseconds or seconds), just in time compilation will lead to compiler lag. There are probably workarounds here, but julia may also not be the right tool for the job.
- If you need to avoid runtime errors at all costs, consider a language like Rust. 
- If the code you need is written in python / fortran / C / C++ / matlab / etc and wrapping it would be more trouble than it's worth. My condolences, but yes, go ahead and use the relevant language. 
- If you want to make something seem user friendly to the scientific community, python certainly seems like the favorite. That said, julia is growing quickly and I'm not sure how a julia code would be received. 
- Here is an interesting discussion on the presence of bugs in newer julia libraries https://itnext.io/why-i-still-recommend-the-julia-programming-language-d2b84a747a5c
- If you know how to use python / fortran / C / C++ / matlab / etc and learning a new language will not help you prepare your presentation or write your paper in the near term. It is clearly okay to use the easiest tool for you, or the one you know how to use. In the long run, however, it is worth considering why we choose to tools we do and if better options exist. 
    - Example: I backed up code on google drive for the first 3 years of my scientific career before getting around to learning git, which in the end took a couple hours. Don't be like me. 

### **F. Type Stability is Very, Very Important**
See also *Julia High Performance* by Avik Sengupta and Alan Edelman
#### Example of type stability in functions

```julia
f(x) = x < 0 ? 0 : x # not type stable (returns an Int64 or typeof(x) depending on whether x < 0)
f2(x) = x < 0 ? zero(x) : x # type stable 
println("Time for type unstable version:")
@btime f.($-.($rand(1000,1000), 0.5));
println("Time for type stable version:")
@btime f2.($-.($rand(1000,1000), 0.5));
println(" ~ 7 x difference!")
```

#### Example of type stability in structs


```julia
abstract type Thing end 
struct Thing1 <: Thing # the type of x is ambiguously defined
    x::Number
end
struct Thing2{T <: Number} <: Thing # type of x is defined by type Thing2{T}
    x::T
end

# define Thing addition
import Base: + 
+(t1::Thing,t2::Thing) = t1.x + t2.x 
    
thing1s = Thing1.(rand(1000,1000))
thing2s = Thing2.(rand(1000,1000))
println("Time for type unstable version:")
@btime thing1s .+ thing1s;
println("Time for type stable version:")
@btime thing2s .+ thing2s;
println(" > 100 x difference!")
```

### For high performance julia code, you do not have to declare types **but you MUST respect type stability**
